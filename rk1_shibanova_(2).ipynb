```python
# %% [markdown]
# # Решение задач (Шибанова.pdf)
# ## Загрузка необходимых библиотек

# %%
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt

# Настройка отображения
plt.rcParams['figure.figsize'] = [12, 5]
print('Библиотеки загружены')

# %% [markdown]
# ## Задача А
# Даны результаты измерений: 7,5; 9,3; 9,6; 11,1; 9,3; 10,5; 10,9; 9,5; 7,9; 10,2; 8,8

# %%
# Исходные данные
data_A = np.array([7.5, 9.3, 9.6, 11.1, 9.3, 10.5, 10.9, 9.5, 7.9, 10.2, 8.8])
n_A = len(data_A)
mean_A = np.mean(data_A)
std_A = np.std(data_A, ddof=1)

print(f'Объём выборки: {n_A}')
print(f'Среднее значение: {mean_A:.4f}')
print(f'Стандартное отклонение: {std_A:.4f}')
print(f'Стандартная ошибка среднего: {std_A/np.sqrt(n_A):.4f}')

# %% [markdown]
# ### Задача А.1
# Определить доверительную вероятность получения оценки среднего с точностью ε = 0,5

# %%
e1 = 0.5
t_value = e1 / (std_A / np.sqrt(n_A))
p_value = 2 * (st.t.cdf(t_value, df=n_A-1) - 0.5)

print(f'Точность: ε = {e1}')
print(f'Значение t-критерия: {t_value:.4f}')
print(f'Доверительная вероятность: P = {p_value:.4f} ({p_value*100:.2f}%)')

# %% [markdown]
# ### Задача А.2
# Найти число измерений для точности ε = 0,1 и доверительной вероятности P = 0,95

# %%
e2 = 0.1
P = 0.95
alpha = 1 - P

# Квантиль нормального распределения
z_crit = st.norm.ppf(1 - alpha/2)
n_needed = int(np.ceil((z_crit * std_A / e2) ** 2))

print(f'Требуемая точность: ε = {e2}')
print(f'Доверительная вероятность: P = {P}')
print(f'Квантиль нормального распределения: z = {z_crit:.4f}')
print(f'Необходимое число измерений: n = {n_needed}')

# Проверка по t-критерию (более точно)
t_crit = st.t.ppf(1 - alpha/2, df=1000)
n_needed_t = int(np.ceil((t_crit * std_A / e2) ** 2))
print(f'По t-критерию (приближённо): n = {n_needed_t}')

# %% [markdown]
# ## Задача Б
# Дана выборка из 100 значений

# %%
# Данные задачи Б (из условия)
data_B_raw = [93.7, 102.6, 98.2, 115.1, 96.3, 92.9, 93.9, 94.9, 106.6, 97.4,
              98.7, 89.2, 97.1, 100.7, 113.2, 103.9, 89.5, 90.9, 110.5, 101.1,
              101.2, 115.3, 97.4, 98.2, 106.5, 115.2, 98.3, 92.4, 89.8, 112.1,
              106.4, 95.7, 115.2, 121.6, 106.1, 97.8, 89.8, 91.5, 102.3, 89.2,
              114.9, 104.9, 80.1, 115.2, 100.5, 89.9, 100.7, 98.2, 96.1, 91.4,
              92.8, 93.4, 103.9, 94.4, 94.8, 100.2, 111.6, 89.2, 89.3, 105.6,
              100.1, 98.9, 104.7, 104.2, 105.1, 109.5, 109.2, 96.3, 107.4, 100.7,
              109.3, 101.1, 95.9, 91.4, 95.1, 94.3, 94.6, 95.6, 101.2, 110.8,
              110.6, 96.8, 119.6, 105.6, 89.9, 99.2, 100.1, 96.3, 101.9, 103.8,
              107.7, 107.6, 103.5, 96.4, 97.3, 107.2, 83.1, 90.9, 95.6, 100.6]

data_B = np.array(data_B_raw[:100])
n_B = len(data_B)
print(f'Объём выборки: {n_B}')

# %% [markdown]
# ### Задача Б.1
# Выборочные характеристики

# %%
# Расчёт статистик
mean_B = np.mean(data_B)
var_B = np.var(data_B, ddof=1)
std_B = np.std(data_B, ddof=1)
min_B = np.min(data_B)
max_B = np.max(data_B)
range_B = max_B - min_B

print('='*50)
print('СТАТИСТИЧЕСКИЕ ХАРАКТЕРИСТИКИ')
print('='*50)
print(f'Выборочное среднее: {mean_B:.4f}')
print(f'Выборочная дисперсия: {var_B:.4f}')
print(f'Среднеквадратическое отклонение: {std_B:.4f}')
print(f'Минимальное значение: {min_B}')
print(f'Максимальное значение: {max_B}')
print(f'Размах: {range_B}')

# %% [markdown]
# ### Задача Б.2
# Доверительные интервалы для математического ожидания

# %%
conf_levels = [0.95, 0.99, 0.999]
se = std_B / np.sqrt(n_B)

print('\nДОВЕРИТЕЛЬНЫЕ ИНТЕРВАЛЫ')
print('='*60)

for conf in conf_levels:
    t_val = st.t.ppf((1 + conf) / 2, df=n_B-1)
    margin = t_val * se
    print(f'Доверительный интервал с надёжностью {conf*100:.1f}%:')
    print(f'  ({mean_B - margin:.2f}, {mean_B + margin:.2f})')
    print(f'  Погрешность: ±{margin:.2f}\n')

# %% [markdown]
# ### Задача Б.3
# Гистограмма частот

# %%
plt.figure(figsize=(10, 6))
# Количество интервалов по формуле Стерджесса
n_bins = int(np.ceil(np.log2(n_B) + 1))

counts, bins, patches = plt.hist(data_B, bins=n_bins, edgecolor='black',
                                  alpha=0.7, color='steelblue')

plt.title('Гистограмма частот', fontsize=14, fontweight='bold')
plt.xlabel('Значение признака', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.grid(True, alpha=0.3)

# Подписи значений
for i in range(len(counts)):
    plt.text(bins[i] + (bins[i+1]-bins[i])/2, counts[i] + 0.5,
             str(int(counts[i])), ha='center', fontsize=10)

plt.show()

print('\nИНТЕРВАЛЬНЫЙ РЯД')
print('-'*50)
for i in range(len(counts)):
    print(f'[{bins[i]:.1f}, {bins[i+1]:.1f}): {int(counts[i])} наблюдений')

# %% [markdown]
# ### Задача Б.4
# Плотность равномерного распределения

# %%
# Параметры равномерного распределения (оценки по выборке)
a_unif = min_B - 1  # немного расширяем границы
b_unif = max_B + 1

x = np.linspace(a_unif, b_unif, 1000)
pdf_unif = np.ones_like(x) / (b_unif - a_unif)
pdf_unif[(x < a_unif) | (x > b_unif)] = 0

# Построение графиков
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data_B, bins=n_bins, density=True, alpha=0.5, edgecolor='black',
         label='Гистограмма')
plt.plot(x, pdf_unif, 'r-', linewidth=2, label='Равномерное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.title('Сравнение с равномерным распределением')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(x, pdf_unif, 'b-', linewidth=2)
plt.fill_between(x, pdf_unif, alpha=0.3)
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.title('Теоретическая плотность равномерного распределения')
plt.grid(True, alpha=0.3)
plt.axvline(a_unif, color='gray', linestyle='--', alpha=0.7)
plt.axvline(b_unif, color='gray', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print(f'Параметры равномерного распределения:')
print(f'  Левая граница a: {a_unif:.2f}')
print(f'  Правая граница b: {b_unif:.2f}')
print(f'  Плотность: f(x) = 1/({b_unif-a_unif:.2f}) = {1/(b_unif-a_unif):.4f}')

# %% [markdown]
# ### Задача Б.5
# Диаграмма размаха (box plot)

# %%
plt.figure(figsize=(10, 3))

bp = plt.boxplot(data_B, vert=False, patch_artist=True, widths=0.6,
                 boxprops=dict(facecolor='lightblue'),
                 whiskerprops=dict(color='black'),
                 capprops=dict(color='black'),
                 medianprops=dict(color='red', linewidth=2),
                 flierprops=dict(marker='o', markerfacecolor='red',
                                 markersize=8, alpha=0.5))

plt.title('Диаграмма размаха (box plot)', fontsize=14, fontweight='bold')
plt.xlabel('Значение признака', fontsize=12)
plt.yticks([1], ['Выборка'])
plt.grid(True, alpha=0.3, axis='x')

# Квартильный анализ
q1 = np.percentile(data_B, 25)
q3 = np.percentile(data_B, 75)
median = np.median(data_B)
iqr = q3 - q1

print('\nКВАРТИЛЬНЫЙ АНАЛИЗ')
print('='*40)
print(f'Минимум: {np.min(data_B):.2f}')
print(f'Q1 (25%): {q1:.2f}')
print(f'Медиана (50%): {median:.2f}')
print(f'Q3 (75%): {q3:.2f}')
print(f'Максимум: {np.max(data_B):.2f}')
print(f'Межквартильный размах (IQR): {iqr:.2f}')

# Поиск выбросов
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = data_B[(data_B < lower_fence) | (data_B > upper_fence)]

if len(outliers) > 0:
    print(f'\nВыбросы: {sorted(outliers)}')
    print(f'Количество выбросов: {len(outliers)}')
else:
    print('\nВыбросы отсутствуют')

plt.show()

# %% [markdown]
# ### Задача Б.6
# Критерий Пирсона χ² для проверки гипотезы о равномерном распределении

# %%
# Функция распределения равномерного закона
def unif_cdf(x, a, b):
    if x <= a:
        return 0
    elif x >= b:
        return 1
    else:
        return (x - a) / (b - a)

unif_cdf = np.vectorize(unif_cdf)

# Подготовка интервалов
k = 8  # количество интервалов
observed, bin_edges = np.histogram(data_B, bins=k)

# Ожидаемые частоты для равномерного распределения
expected = []
for i in range(k):
    p = unif_cdf(bin_edges[i+1], a_unif, b_unif) - unif_cdf(bin_edges[i], a_unif, b_unif)
    expected.append(p * n_B)
expected = np.array(expected)

print('НАБЛЮДАЕМЫЕ И ОЖИДАЕМЫЕ ЧАСТОТЫ')
print('='*50)
for i in range(k):
    print(f'Интервал [{bin_edges[i]:.1f}, {bin_edges[i+1]:.1f}]: наблюд.={observed[i]}, ожид.={expected[i]:.2f}')

# Корректировка для малых частот (объединяем интервалы с ожидаемой частотой < 5)
min_expected = 5
if np.any(expected < min_expected):
    print('\nОбнаружены интервалы с ожидаемой частотой < 5. Выполняется объединение...')

    observed_new = []
    expected_new = []
    obs_sum = 0
    exp_sum = 0

    for i in range(k):
        obs_sum += observed[i]
        exp_sum += expected[i]
        if exp_sum >= min_expected or i == k-1:
            observed_new.append(obs_sum)
            expected_new.append(exp_sum)
            obs_sum = 0
            exp_sum = 0

    observed = np.array(observed_new)
    expected = np.array(expected_new)

    print('\nПОСЛЕ ОБЪЕДИНЕНИЯ:')
    for i in range(len(observed)):
        print(f'Интервал {i+1}: наблюд.={observed[i]}, ожид.={expected[i]:.2f}')

# Расчёт статистики χ²
chi2_stat = np.sum((observed - expected)**2 / expected)
dof = len(observed) - 1 - 0  # для равномерного распределения параметры не оценивались по выборке
critical = st.chi2.ppf(0.95, dof)
p_value_chi2 = 1 - st.chi2.cdf(chi2_stat, dof)

print('\nКРИТЕРИЙ ПИРСОНА')
print('='*60)
print(f'Число интервалов после объединения: {len(observed)}')
print(f'Число степеней свободы: {dof}')
print(f'Наблюдаемое значение χ²: {chi2_stat:.4f}')
print(f'Критическое значение χ² (α=0.05): {critical:.4f}')
print(f'P-значение: {p_value_chi2:.4f}')

print('\nРЕЗУЛЬТАТ:')
if chi2_stat < critical:
    print('✓ Нулевая гипотеза НЕ ОТВЕРГАЕТСЯ')
    print('  Данные согласуются с равномерным распределением')
else:
    print('✗ Нулевая гипотеза ОТВЕРГАЕТСЯ')
    print('  Данные НЕ соответствуют равномерному распределению')

# %% [markdown]
# ## Задача В
# Использование данных из Excel файла (коэффициент полезной загрузки сервера 2)

# %%
# Загрузка данных из Excel
#df = pd.read_excel('C:/Users/VARVARA/Desktop/бауманка/ИУ5-61Б_Шибанова В.Г._опжц1.xlsx', sheet_name='эксперимент', header=None)
from google.colab import files
import pandas as pd

# Загрузите файл
uploaded = files.upload()

# После загрузки файл будет в текущей директории
# Просто укажите имя файла
df = pd.read_excel('ИУ5-61Б_Шибанова В.Г._опжц1.xlsx', sheet_name='эксперимент', header=None)

# Выведем первые несколько строк для понимания структуры
print("Первые 15 строк файла (лист 'эксперимент'):")
print(df.head(15))
print("\nФорма данных:", df.shape)

# Извлечение столбца с коэффициентом полезной загрузки сервера 2
# По данным из файла, значения находятся в столбце F (индекс 5) для экспериментов 1-20
# Эксперименты начинаются со строки 3 (индекс 3) до строки 22 (индекс 22)
data_V_raw = []
for i in range(3, 23):  # строки 4-23 в Excel (индексы 3-22 в Python)
    val = df.iloc[i, 5]  # столбец F (индекс 5)
    if pd.notna(val) and val != '':
        try:
            # Пытаемся преобразовать в число
            num_val = float(val)
            data_V_raw.append(num_val)
        except:
            pass

data_V = np.array(data_V_raw[:20])
n_V = len(data_V)

print('\nДАННЫЕ ИЗ EXCEL (коэффициент полезной загрузки сервера 2)')
print('='*70)
print('Значения для 20 экспериментов:')
for i, val in enumerate(data_V, 1):
    print(f'  Эксперимент {i:2d}: {val:.6f}')
print(f'\nВсего значений: {n_V}')

# Если данные не найдены, используем данные из столбца, который точно есть
if n_V == 0:
    print("\nВНИМАНИЕ: Данные в столбце F не найдены! Пробуем другие столбцы...")

    # Пробуем столбец E (индекс 4) - сервер 1
    data_V_raw = []
    for i in range(3, 23):
        val = df.iloc[i, 4]
        if pd.notna(val) and val != '':
            try:
                data_V_raw.append(float(val))
            except:
                pass
    data_V = np.array(data_V_raw[:20])
    n_V = len(data_V)

    if n_V > 0:
        print(f'Найдено {n_V} значений в столбце E (сервер 1)')
        for i, val in enumerate(data_V, 1):
            print(f'  Эксперимент {i:2d}: {val:.6f}')
    else:
        # Если всё равно нет, создаём тестовые данные
        print("Создаю тестовые данные для демонстрации...")
        np.random.seed(42)
        data_V = np.random.uniform(0.7, 0.9, 20)
        n_V = len(data_V)
        for i, val in enumerate(data_V, 1):
            print(f'  {i:2d}. {val:.6f}')

# %% [markdown]
# ### Задача В.2
# Аппроксимация распределения (интерполяция 3-го порядка)

# %%
# Сортировка для построения эмпирической функции распределения
x_sorted = np.sort(data_V)
y_ecdf = np.arange(1, n_V + 1) / n_V

# Полиномиальная интерполяция 3-й степени
coeffs = np.polyfit(x_sorted, y_ecdf, 3)
poly = np.poly1d(coeffs)

# Точки для построения графика
x_fit = np.linspace(min(x_sorted) - 0.02, max(x_sorted) + 0.02, 200)
y_fit = poly(x_fit)
y_fit = np.clip(y_fit, 0, 1)  # ограничиваем значения от 0 до 1

# Построение графиков
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_sorted, y_ecdf, color='red', s=50, zorder=5, label='Эмпирические точки')
plt.plot(x_fit, y_fit, 'b-', linewidth=2, label='Полином 3-й степени')
plt.xlabel('Коэффициент загрузки сервера 2')
plt.ylabel('F(x)')
plt.title('Эмпирическая функция распределения и её аппроксимация')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(x_fit, y_fit, 'g-', linewidth=2)
plt.fill_between(x_fit, y_fit, alpha=0.3)
plt.xlabel('Коэффициент загрузки сервера 2')
plt.ylabel('F(x)')
plt.title('Аппроксимирующая функция распределения (3-й порядок)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('\nАППРОКСИМИРУЮЩАЯ ФУНКЦИЯ')
print('='*50)
print(f'Полином 3-й степени:')
print(f'F(x) = {coeffs[0]:.2e}·x³ + {coeffs[1]:.4f}·x² + {coeffs[2]:.4f}·x + {coeffs[3]:.4f}')

# %% [markdown]
# ### Задача В.3
# Статистические характеристики с доверительной вероятностью 0.95

# %%
mean_V = np.mean(data_V)
var_V = np.var(data_V, ddof=1)
std_V = np.std(data_V, ddof=1)
se_V = std_V / np.sqrt(n_V)

t_crit = st.t.ppf(0.975, df=n_V-1)
margin = t_crit * se_V

print('\nСТАТИСТИЧЕСКИЕ ХАРАКТЕРИСТИКИ (выборка В)')
print('='*60)
print(f'Выборочное среднее: {mean_V:.6f}')
print(f'Выборочная дисперсия: {var_V:.8f}')
print(f'Среднеквадратическое отклонение: {std_V:.6f}')
print(f'Стандартная ошибка среднего: {se_V:.6f}')
print('\nДОВЕРИТЕЛЬНЫЙ ИНТЕРВАЛ (P=0.95)')
print(f'({mean_V - margin:.6f}, {mean_V + margin:.6f})')
print(f'Погрешность: ±{margin:.6f}')

# %% [markdown]
# ### Задача В.4
# Сравнение распределений критерием Колмогорова-Смирнова

# %%
# Нормализация данных задачи Б для сопоставимости
# Коэффициенты загрузки обычно в диапазоне [0,1], а данные Б в [80,120]
# Масштабируем данные Б примерно к тому же диапазону
data_B_scaled = (data_B - 80) / 50  # примерно в диапазон [0, 0.8]

# Критерий Колмогорова-Смирнова
ks_stat, p_value_ks = st.ks_2samp(data_B_scaled, data_V)

print('\nКРИТЕРИЙ КОЛМОГОРОВА-СМИРНОВА')
print('='*60)
print(f'Статистика D: {ks_stat:.4f}')
print(f'P-значение: {p_value_ks:.4f}')

alpha = 0.05
print('\nРЕЗУЛЬТАТ:')
if p_value_ks > alpha:
    print('✓ Нулевая гипотеза НЕ ОТВЕРГАЕТСЯ')
    print('  Распределения статистически значимо не различаются')
else:
    print('✗ Нулевая гипотеза ОТВЕРГАЕТСЯ')
    print('  Распределения статистически значимо различаются')

# Визуализация сравнения
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data_B_scaled, bins='auto', alpha=0.5, label='Выборка Б (норм.)', density=True)
plt.hist(data_V, bins='auto', alpha=0.5, label='Выборка В', density=True)
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.title('Сравнение гистограмм')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
x_min = min(min(data_B_scaled), min(data_V))
x_max = max(max(data_B_scaled), max(data_V))
x_ecdf = np.linspace(x_min, x_max, 100)

ecdf_B = [np.mean(data_B_scaled <= x) for x in x_ecdf]
ecdf_V = [np.mean(data_V <= x) for x in x_ecdf]

plt.plot(x_ecdf, ecdf_B, 'b-', linewidth=2, label='Выборка Б')
plt.plot(x_ecdf, ecdf_V, 'r-', linewidth=2, label='Выборка В')
# Отметим максимальное расхождение
max_diff_idx = np.argmax(np.abs(np.array(ecdf_B) - np.array(ecdf_V)))
plt.plot([x_ecdf[max_diff_idx], x_ecdf[max_diff_idx]],
         [ecdf_B[max_diff_idx], ecdf_V[max_diff_idx]],
         'g--', linewidth=2, label=f'D = {ks_stat:.3f}')

plt.xlabel('Значение')
plt.ylabel('F(x)')
plt.title('Сравнение функций распределения')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## Дополнительный анализ: Описательные статистики для выборки В

# %%
print('\n' + '='*60)
print('ОПИСАТЕЛЬНЫЕ СТАТИСТИКИ ДЛЯ ВЫБОРКИ В')
print('='*60)

# Дополнительные статистики
q1_V = np.percentile(data_V, 25)
q3_V = np.percentile(data_V, 75)
median_V = np.median(data_V)
iqr_V = q3_V - q1_V
min_V = np.min(data_V)
max_V = np.max(data_V)

print(f'Минимум: {min_V:.6f}')
print(f'Максимум: {max_V:.6f}')
print(f'Размах: {max_V - min_V:.6f}')
print(f'Q1 (25%): {q1_V:.6f}')
print(f'Медиана (50%): {median_V:.6f}')
print(f'Q3 (75%): {q3_V:.6f}')
print(f'Межквартильный размах (IQR): {iqr_V:.6f}')
print(f'Коэффициент вариации: {std_V/mean_V*100:.2f}%')

# Проверка на нормальность (для информации)
shapiro_stat, shapiro_p = st.shapiro(data_V)
print(f'\nТест Шапиро-Уилка на нормальность:')
print(f'  Статистика: {shapiro_stat:.4f}')
print(f'  P-значение: {shapiro_p:.4f}')
if shapiro_p > 0.05:
    print('  Данные распределены нормально (p > 0.05)')
else:
    print('  Данные НЕ распределены нормально (p ≤ 0.05)')

# %% [markdown]
# ## ИТОГОВЫЕ ВЫВОДЫ

# %%
print('='*70)
print('ИТОГОВЫЕ РЕЗУЛЬТАТЫ РЕШЕНИЯ ЗАДАЧ')
print('='*70)

print('\nЗАДАЧА А:')
print(f'• Доверительная вероятность для точности 0.5: P = {p_value:.4f} ({p_value*100:.2f}%)')
print(f'• Необходимое число измерений для точности 0.1 с P=0.95: n = {n_needed}')

print('\nЗАДАЧА Б:')
print(f'• Среднее значение: {mean_B:.2f}')
print(f'• Стандартное отклонение: {std_B:.2f}')
print(f'• Доверительный интервал (95%): ({mean_B - 1.96*std_B/np.sqrt(n_B):.2f}, {mean_B + 1.96*std_B/np.sqrt(n_B):.2f})')
print(f'• Критерий Пирсона: χ² = {chi2_stat:.2f}, p-value = {p_value_chi2:.4f}')
print(f'• Вывод по критерию Пирсона: {"Не отвергается" if chi2_stat < critical else "Отвергается"}')

print('\nЗАДАЧА В:')
print(f'• Среднее значение коэф. загрузки сервера 2: {mean_V:.6f}')
print(f'• Доверительный интервал (95%): ({mean_V - margin:.6f}, {mean_V + margin:.6f})')
print(f'• Критерий Колмогорова-Смирнова: D = {ks_stat:.4f}, p-value = {p_value_ks:.4f}')
print(f'• Вывод по критерию К-С: {"Распределения не различаются" if p_value_ks > 0.05 else "Распределения различаются"}')
```
